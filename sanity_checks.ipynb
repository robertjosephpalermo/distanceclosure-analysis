```python
import zipfile
from glob import glob
```

```python
fcount = 0
for zfile in glob('LOCAL/MRI Clouds Diffusion/Diffusion/*.zip'):
    with zipfile.ZipFile(zfile, 'r') as z:
        print(zfile)
        for fname in z.namelist():
            split_fname = fname.split('/')
            if len(split_fname) == 3 and split_fname[-1] != '':
                fcount += 1
```
