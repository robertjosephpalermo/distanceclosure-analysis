```python
import pandas as pd
import zipfile
from glob import glob
import networkx as nx

from sklearn.preprocessing import MinMaxScaler
from populate_data import single_use_backbone_output
```

```python
import os
def ensurePathExists(filepath):
    """Given a file with path, ensures a path exists by creating a directory if needed. """
    if not os.path.exists(os.path.dirname(filepath)):
        os.makedirs(os.path.dirname(filepath))
```

```python
scaler = MinMaxScaler(feature_range=(1e-15, 1-1e-15), clip=True) # type: ignore
```

```python
for zfile in glob('../LOCAL/MRI Clouds Diffusion/Diffusion/*.zip'):
    with zipfile.ZipFile(zfile, 'r') as z:
        for fname in z.namelist():
            split_fname = fname.split('/')
            if len(split_fname) == 3 and split_fname[-1] != '':
                with z.open(fname) as f:
                    df = pd.read_csv(f, header=None, names=['source', 'target', 'counts'], sep=' ')
                    df['weight'] = scaler.fit_transform(df[['counts']])
                    df['distance'] = 1./df['weight'] - 1
                    
                    G = nx.from_pandas_edgelist(df, edge_attr='distance')
                    
                    wFile = 'BrainBackbone/' + fname
                    ensurePathExists(wFile)
                    single_use_backbone_output(graph=G, edge_attribute="distance", output_file_name=wFile)
                    
                    break
    
    break
```
